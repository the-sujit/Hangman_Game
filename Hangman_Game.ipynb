```python
# Hangman Game
import random
import string

## file name is 'words.txt'

def load_words():
   # opening a file reading each line and spliting it into strings and returning each string into wordlist 
    wordlist = open('words.txt', 'r').readline().split()
    return wordlist

def choose_word(wordlist):
    return random.choice(wordlist)

def concat(strn, letter):
    if strn == None:
         strn = letter + ' '
    else :
        strn = strn + letter + ' '
    
    return strn

## function which returns guessed word
def get_guessed_word(secret_word, letter_guessed):

    strn = None
    for letter in secret_word: # running chances 
        if letter in letter_guessed:
            strn = concat(strn , letter)

        else :
            strn = concat(strn, '_')
    return strn

## function that tells whether the word has been guessed or not

def is_word_guessed(secret_word, letter_guessed):    
    get_guessed_word(secret_word, letter_guessed)
    return secret_word == get_guessed_word(secret_word, letter_guessed).replace(' ','')

## function which gives all the alphabets other than guessed ones

def get_available_letters(secret_word, letter_guessed):
    strn = None
    for letter in string.ascii_lowercase:
        if letter not in letter_guessed:
            strn = concat(strn, letter)
        else :
            continue
      
    return strn
# hangman game

def hangman():
    
    worldlist = load_words()
    secret_word = choose_word(wordlist)
    print("secret word has", len(secret_word),"letters")
    warnings = 3
    print("you have", len(secret_word), "chances left")
    letter_guessed = []
    
    g = len(secret_word)
    while g > 0:
            guess_letter = input("enter guess letter:")
            letter_guessed.append(guess_letter)
            
            if guess_letter in secret_word:
                g = g
                
            else :
                g = g-1
            
            print(get_guessed_word(secret_word, letter_guessed))
            print(get_available_letters(secret_word, letter_guessed))
            print("You have", g, "chances left")
            
            if is_word_guessed(secret_word, letter_guessed):
                print("You won! The correct word was", secret_word)
                break
            elif g!=0:
                continue
            else :
                print("You lost! The correct word was", secret_word)
                break

# function which tells whether guessed word can be given word

def match_with_gaps(guessed_word, word):
        if len(word) != len(guessed_word):
            return False
        else :
            for i in range(len(guessed_word)):
                if guessed_word[i] == '_':
                    continue
                    
                elif guessed_word[i] != word[i]:
                    return False
                    break
                else :
                    continue
            return True

# function which return possible matches of guessed word from wordlist

def show_possible_matches(guessed_word):
    possible_matches = []
    for word in wordlist:
        if match_with_gaps(guessed_word, word):
            possible_matches.append(word)
        else :
            continue
    return possible_matches

# hangman game which shows possible matches of guessed word after every guessed letter

def hangman_with_hints():
    
    wordlist = load_words()
    secret_word = choose_word(wordlist)
    print("secret word has", len(secret_word),"letters")
    warnings = 3
    print("you have", len(secret_word), "chances and", warnings,"warnings left")
    letter_guessed = []

    g = len(secret_word)
    warnings = 3
    while g > 0:
        
        guess_letter = input("enter guess letter:")
        if guess_letter in string.ascii_lowercase:
            warnings = warnings
            letter_guessed.append(guess_letter)
            
            guessed_word = get_guessed_word(secret_word, letter_guessed).replace(' ','')
            print(show_possible_matches(guessed_word))
            
            if guess_letter in secret_word:
                g = g
                print("Bang On! Thats a good guess. ", get_guessed_word(secret_word, letter_guessed))
            
            else :
                g = g-1
                print("Oops! Not a good guess. ", get_guessed_word(secret_word, letter_guessed))
            
            
            
            print("Available Letters:", get_available_letters(secret_word, letter_guessed))
            print("You have", g, "chances left")
            
            if is_word_guessed(secret_word, letter_guessed):
                print("You won! The correct word was", secret_word)
                print("Your score is", int(g)*len(set(secret_word)))
                break
            elif g!=0:
                continue
            else :
                print("You lost! The correct word was", secret_word)
                break
        else :
               print("That is not a valid lowercase alphabet")
               if warnings > 0:
                   warnings = warnings - 1
               else :
                   g = g-1
               continue
```
